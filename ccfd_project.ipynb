# Input Splitting

```python
X = df.drop(columns=['Class'], axis= 1)
y = df['Class']
```

```python
# standard scaling 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_scaler = sc.fit_transform(X)
```

```python
x_scaler[-1]
```

# Model Training

```python
# train test split 
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
x_train , x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.25, random_state=42, stratify=y )
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

# training 
model.fit(x_train, y_train)

# testing 
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

>  Interpretation: This score indicates a moderate balance between precision and recall. The model is reasonably effective but has room for improvement in either precision, recall, or both.
> Possible Issues: The model might be missing some fraudulent transactions (lower recall) or incorrectly flagging non-fraudulent transactions as fraudulent (lower precision).

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

# training 
model.fit(x_train, y_train)

# testing 
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

> Interpretation: This score is significantly better, indicating a good balance between precision and recall. The model is more effective at identifying fraudulent transactions while minimizing false positives and false negatives.
> Improvement: This model has improved either precision, recall, or both compared to the first model, making it more reliable for fraud detection.

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_jobs=-1)

# training 
model.fit(x_train, y_train)

# testing 
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

> Interpretation: This is the highest F1 score among the three, indicating the best balance between precision and recall. The model is highly effective at identifying fraudulent transactions with minimal errors.
> Best Performance: This model performs the best, suggesting it has the most optimal balance between identifying true positives and minimizing false positives.

# Class Imbalancement

```python
# hint - use combination of under sampling and over sampling 
# balance the class with equal distribution 
from imblearn.over_sampling import SMOTE
over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train , y_train)
```

```python
# print(y_smote.value_counts())
sns.countplot(x=y_smote)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution after SMOTE')
plt.show()
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_jobs=-1)

# training 
model.fit(x_smote, y_smote)

# testing 
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```
